```python
df['search_terms_list'] = [ast.literal_eval(x) for x in df['search_terms'] ]
df.head()
```

```python
def remove_bracket(list_x):
    for x in list_x:
        x = x.strip()
        if '{' in x:
            x = x.replace('{','')
            list_x[0] = x
        
        if '}' in x:
            print(x)
            x = x.replace('}','')
            print(x)
            list_x[-1] = x
    
    return list_x
```

```python
df['search_terms_list'] = df['search_terms_list'].apply(lambda x: list(x))
df.head()
```

```python
def strip_list(x_list):
    return [x.strip() for x in x_list]
```
